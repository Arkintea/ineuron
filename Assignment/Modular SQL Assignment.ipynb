```python
import logging
logging.basicConfig(filename='map_reduce_filter_assignment.log', 
                    level=logging.DEBUG, 
                    format='%(asctime)s %(levelname)s %(message)s', 
                   filemode='w')
import csv

class CountAlphabet:
    def __init__(self, data):
        """instance variable for the file"""
        logging.info('create the instance variable for the file')
        self.data = data

    def alphabet_counter(self):
        """count of every letter in a word"""
        logging.info('execute the alphabet_counter function')
        try:
            with open(f'{self.data}.txt', 'r', encoding='utf-8') as f:
                enron_data = csv.reader(f, delimiter='\n')
                lst = []
                for i in enron_data:
                    tup = (i[0], len(i[0]))
                    lst.append(tup)
                return lst
        except Exception as e:
            logging.error('Execution error occured', e)
        else:
            print('Try block successfully completed')
        finally:
            print('Done....')
```

```python
obj = CountAlphabet('enron')
obj.alphabet_counter()
```

```python
import logging
logging.basicConfig(filename='word_filter_assignment.log', 
                    level=logging.DEBUG, 
                    format='%(asctime)s %(levelname)s %(message)s', 
                    filemode='w')
import csv

class word_filter:
    def __init__(self, lst):
        """instance variable for the file"""
        logging.info('create the instance variable for the file')
        self.lst = lst
    
    def filter_word(self):
        """filter the file to return just words (alphabets) eliminating non alphabet characters"""
        logging.info('execute the filter word function')
        try:
            import re
            with open(f'{self.lst}.txt', 'r', encoding='utf-8') as f:
                enron_data = csv.reader(f, delimiter='\n')
                lst = []
                for i in enron_data:
                    tup = (i[0])
                    lst.append(tup)
                lst

                new_lst = []
                for i in lst:
                    new = re.sub('[^a-zA-Z]', '', str(i))
                    new_lst.append(new)
                new_lst

                xet = []
                for i in new_lst:
                    if i != '':
                        xet.append(i)
                return xet
            
        except Exception as e:
            logging.error('Execution error occured', e)
        else:
            print('Try block successfully completed')
        finally:
            print('Done....')
```

```python
wrek = word_filter('pubmed')
wrek.filter_word()
```

```python
import logging
logging.basicConfig(filename='merged_file.log', 
                    level=logging.DEBUG, 
                    format='%(asctime)s %(levelname)s %(message)s', 
                    filemode='w')
import csv

class file_merger(word_filter):
    def __init__(self, files):
        """instance variable for the file"""
        logging.info('create the instance variable for the file')
        self.files = files
    
    def merge_file(self):
        """merge files from a list and insert into sqlite database"""
        logging.info('execute the merge file function')
        try:            
            file = self.files
            merged = []
            for i in file:
                wrek = word_filter(i)
                merged.append(wrek.filter_word())
            merged_tuple = [tuple(i) for i in merged]

            #merge all files together as a single list 
            rat = list(zip(*merged_tuple))
            
            import sqlite3
            db = sqlite3.connect('Assignment.db')
            c = db.cursor()
            c.execute('create table mapping(File_1 text, File_2 text, File_3 text, File_4 text, File_5 text)')

            for i in rat:
                #c.execute(f'insert into mapping values({str(i[0])})')
                c.execute("insert into mapping values('{}','{}','{}','{}','{}')".format(str(i[0]), str(i[1]), str(i[2]), str(i[3]), str(i[4])))

            data = c.execute('select * from mapping')
            return data

            #db.commit()
            #db.close()
            
        except Exception as e:
            logging.error('Execution error occured', e)
        else:
            print('Try block successfully completed')
        finally:
            print('Check other methods....')
```
